# Optimization

```python
X = np.linspace(0,1,1000)

XX = np.sin((X-0.5)*np.pi)/2 +0.5

plt.plot(X,np.sin((XX-0.5)*np.pi)/2 +0.5)
plt.plot(X,np.sin((X-0.5)*np.pi)/2 +0.5)
plt.plot(X,X)
```

```python
XX = np.sin((X-0.5)*np.pi)/2 +0.5
XXX = np.sin((XX-0.5)*np.pi)/2 +0.5

plt.plot(X,(XX*2-1)*abs(XX*2-1)**0.2/2+0.5)
plt.plot(X,X)
```

```python
res = pd.read_csv("RForest_sample_submission_ml2.csv")

for lab in labels:
    X = res[lab]
#     X = np.sin((X-0.5)*np.pi)/2 +0.5
    X = (X*2-1)*abs(X*2-1)**0.2/2+0.5
    X *= 0.99999
    X += 0.000005
    res[lab] = X
```

```python
res.to_csv('RF2 with modifications', index=False)
```
